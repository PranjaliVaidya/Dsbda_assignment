```python
def Mean(variable):
    s=df[variable].sum()
    c=df[variable].count()
    m=s/c
    return m
    
```

```python
print('mean :',Mean('reading score'))
```

```python
def Median(variable):
    s=df[variable].count()
    mid_index=math.floor(s/2)
    median=np.sort(df[variable])[mid_index]
    return median
    
```

```python
print("median:",Median('reading score'))
```

```python
def std_deviation(variable):
    s=df[variable].sum()
    c=df[variable].count()
    m=s/c
    sigma_val=[]
    for x in range(c):
        sigma=(df[variable][x]-m)**2
        sigma_val.append(sigma)
    std=np.sqrt(sum(sigma_val)/c)
    return std
        
```

```python
print("std_deviation:",std_deviation('reading score'))
```

```python
def Min(variable):
    min=np.sort(df[variable])[0]   #for max=[-1]
    return min
```
